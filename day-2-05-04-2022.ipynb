```python
var3 = np.empty((2,3))
var3[0][0] = 9
var3[0][1] = 2.36
var3[0][2] = 5.6
var3[1][0] = 0.9
var3[1][1] = 1.32
var3[1][2] = 56
print(var3)
print("The dimension of the array ...")
print(var3.shape)
print(var3[1])
print(var3[[0, 1]])
print(var3[:, 2])
print(var3[:, [1, 2]])
print(var3[1][2])

print("Transpose of the matrix....")
print(np.transpose(var3))
print("Re-Shaped the array....")
print(var3.reshape(3,2))
```

```python
Arr_1= np.empty((2,3))
Arr_1[0][0] = 1
Arr_1[0][1] = 2
Arr_1[0][2] = 3
Arr_1[1][0] = 4
Arr_1[1][1] = 5
Arr_1[1][2] = 6

Arr_2 = np.empty((1,3))
Arr_2[0][0] = 10
Arr_2[0][1] = 20
Arr_2[0][2] = 30

Arr_Con = np.concatenate((Arr_1, Arr_2), axis = 0)
print(Arr_Con)

print("The min value in the concatenate matrix is: ")
print(Arr_Con.min())
print("The max value in the concatenate matrix is: ")
print(Arr_Con.max())
print("Cumulative sum of the values stored in an array")
print(Arr_Con.cumsum())
print("Mean or average value stored in an array")
print(Arr_Con.mean())
print("standard deviation of values stored in an array")
print(Arr_Con.std())
```

```python
arr2 = np.array([1.1,2,2,10,20,30,50,5.5,8])
print(arr2)
print(np.around(arr2))
print(np.around(arr2,decimals=5))
print(np.floor(arr2))
print(np.ceil(arr2))
```

```python
arr = np.array([[8,2,1],[14,5,26]])
print(np.sort(arr))
print("The values with axis 0....")
print(np.sort(arr, axis = 0))
print("The values with axis 1....")
print(np.sort(arr, axis = 1))
print("The values in the matrix which are greater than 5.")
print(arr[np.where(arr>5)])
```

```python
arr = np.array([[8,2,1],[14,5,26],[10,56,1]])
print("The mean of the array is: ")
print(np.mean(arr))
print("The mean of the array with axis 0 is: ")
print(np.mean(arr,axis=0))
print("The mean of the array with axis 1 is: ")
print(np.mean(arr,axis=1))
print("The S.D of the array with is: ")
print(np.std(arr))
print(np.var(arr))
```
